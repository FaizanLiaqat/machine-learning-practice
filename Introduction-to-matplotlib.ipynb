# Two options for subplots

```python
# Suplot option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#nrows and ncols defines the dimension of figure 2x2 = four axis
# ax1 top left, ax2 top right, ax3 bottom left, ax4 bottom right

# Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

# Now we use the OO method

```python
heart_disease.head()
```

```python
# creating subset of our data 
# looking at the data of patient who are over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# creating scatter plot between age and chlosterol 
# using pandas
# Plot a scatter plot of 'age' vs 'chol' with points colored by the 'target' variable
# The 'c' parameter is used to set the color of the points based on the values in the 'target' column
# This helps to visually distinguish between different groups in the 'target' variable
over_50.plot(kind= "scatter",x="age",y="chol",c="target");
```

```python
# Now using OO + Pyplot hybrid method
fig,ax=plt.subplots(figsize=(10,6))
# ax=ax tells that we are using OO method 
# and also tells that we want to plot our data on that axis(ax for short)
over_50.plot(kind="scatter", x="age",y="chol",c="target",ax=ax);
# changing the x-axis range
ax.set_xlim([45,100]);
```

```python
## 00 method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
# Create a scatter plot with 'age' on the x-axis and 'chol' on the y-axis
# Color the points based on the 'target' column
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")

# Add a legend to the scatter plot
# The legend_elements() method generates handles and labels for the legend based on the 'c' parameter in the scatter plot
# The * operator unpacks these handles and labels into the ax.legend() method
# This will create a legend using the unique values in the 'target' column
# The title parameter sets the title of the legend to "Target"
ax.legend(*scatter.legend_elements(), title="Target");

# add a horizontal line representing the Cholesterol mean
ax.axhline(y=over_50["chol"].mean(),linestyle='--');
```

```python
# creating a subplot chol vs maxheartrate
fig, ax= plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["thalach"],y=over_50["chol"],c=over_50["target"]);

ax.set(title="Heart disease against Cholesterol and Max Heart Rate",xlabel="Max Heart Rate", ylabel="Cholesterol");

ax.legend(*scatter.legend_elements(),title="Target");

ax.axhline(y=over_50["chol"].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
# we are going to create single figure
# but it will have two axis
# sharex=True tells that our subplots have same x-axis values
fig,(ax0,ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

# add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
# customize ax0
# we have sharex=True so no need to write xlabel for ax0 only need to do for ax1
ax0.set(title="Heart Disease and Cholesterol Levels" ,ylabel="Cholesterol");
# add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");
# add a mean line
ax0.axhline(y=over_50["chol"].mean(),linestyle='--');

# add data to ax1
scatter1=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"]);
# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate");
# add a legend to ax1
ax1.legend(*scatter1.legend_elements(),title="Target");
# add a mean line
ax1.axhline(y=over_50["thalach"].mean(), linestyle='--');

# add a title to entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
# Customizing the y and x axis limitation (remove the left mis-alignment)

# Subplot of chol, age, thalach
# we are going to create single figure
# but it will have two axis
# sharex=True tells that our subplots have same x-axis values
fig,(ax0,ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

# add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter");
# customize ax0
# we have sharex=True so no need to write xlabel for ax0 only need to do for ax1
ax0.set(title="Heart Disease and Cholesterol Levels" ,ylabel="Cholesterol");
# Customizing the x axis limits
ax0.set_xlim(50,80)
# add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");
# add a mean line
ax0.axhline(y=over_50["chol"].mean(),linestyle='--');

# add data to ax1
scatter1=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter");
# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate");
# change ax1 x axis limit
ax1.set_xlim(50,80)
# changing ax1 y axis limit
ax1.set_ylim(60,200)
# add a legend to ax1
ax1.legend(*scatter1.legend_elements(),title="Target");
# add a mean line
ax1.axhline(y=over_50["thalach"].mean(), linestyle='--');

# add a title to entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
